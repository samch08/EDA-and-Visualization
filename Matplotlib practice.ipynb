```python
from matplotlib import style
style.use("ggplot")

x=[3,4,5]
y=[7,5,9]

x1=[5,8,10]
y1=[12,15,9]
#putting data,color,label and linwidth as argument
plt.plot(x,y,'r', label="line1",linewidth='5')
plt.plot(x1,y1,'b', label="line2",linewidth='5')
plt.title("custom graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
#style.use("ggplot")
#defining list
x=['Sambita','Moumita','Ankita']
y=[7,5,9]

x1=['Sambita1','Moumita1','Ankita1']
y1=[12,15,9]
#putting data,color,label and linwidth as argument
plt.bar(x,y,color='r', label="line1",linewidth='5')
plt.bar(x1,y1,color='g', label="line2",linewidth='5')
plt.title("Custom Bar graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```
